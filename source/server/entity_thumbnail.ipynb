```python
import json
import cv2
import os
import random
from shutil import copyfile
```

```python
def make_entity_thumbnail_folder(root_dir, save_dir):
    list_dir = os.listdir(root_dir)
    list_dir.remove('.DS_Store')
    for folder in list_dir:
        folder_path = os.path.join(root_dir, folder)
        filenames = os.listdir(folder_path)
        num = random.randint(0, len(filenames)-1)
        while filenames[num] == '.DS_Store':
            num = random.randint(0, len(filenames)-1)
        thumb_img_path = os.path.join(folder_path, filenames[num])
        os.mkdir(os.path.join(save_dir, folder))
        copyfile(thumb_img_path, f'./{save_dir}/{folder}/{filenames[num]}')
```

```python
root_dir = './pre_DB12/'
save_dir = './server/static'

make_entity_thumbnail_folder(root_dir, save_dir)
```

```python
def thumbnail_info_as_array(folder):
    thumbnail_array = []
    entities = os.listdir(folder)
    entities.remove('.DS_Store')
    for entity in entities:
        sub_dict = {}
        filename = os.listdir(os.path.join(folder, entity))[0]
        directory = os.path.join(entity,filename)
        sub_dict['name'] = entity
        sub_dict['thumb_dir'] = directory
        thumbnail_array.append(sub_dict)
    return thumbnail_array
```

```python
thumbnail_dict = thumbnail_info_as_array('./server/static/')
thumbnail_dict
```

```python
with open('./entity_thumnail.json', 'w') as f:
    out = json.dumps(thumbnail_dict, ensure_ascii=False)
    f.write(out)
```
