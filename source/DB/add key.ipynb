```python
import json
```

```python
with open('./tong.json') as f:   #파일 불러오기
    file = json.load(f)
```

```python
file
```

```python
type(file)
```

```python
for k in range(len(file)):    #key값 지정해주기
    file[k]['key'] = '%05d'%k
```

```python
file    #확인
```

```python
#다시 저장
out = json.dumps(file, ensure_ascii=False)
with open('./tong_with_key.json', "w", encoding="utf8") as f:
    f.write(out)
```
